```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
height = [122,222,421,123]
width = [112,232,233,211]

plt.plot(width,height)
```

```python
plt.figure(figsize=(10,5))
plt.plot(height,'g--')
plt.plot(width,'x')

plt.legend(labels = ['height','width'],  loc = 9)
plt.xticks(ticks=[0,1,2,3],labels=['x','y','z','a'])
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(6,6),sharey= True)
ax[0].plot(width)
ax[1].plot(height)
ax[0].set_title('sdsd')
ax[1].set_title('sdsd1')

ax[0].set_xticks([0,1,2,3])
ax[1].set_xticks([1,2,3,4])
ax[1].set_xticklabels(['a','b','c','d'])
```

```python
fileData = pd.read_csv('/Users/aletheadevelopmentteam/Downloads/Data_Visualization_with_Matplotlib/bigmart_data.csv')
```

```python
fileData = fileData.dropna()
fileData.head()
```

```python
item_mean = fileData.groupby('Item_Type').Item_MRP.mean()
```

```python
plt.figure(figsize=(20,8),dpi=200, clear=True)
plt.xticks(labels=item_mean.index, ticks = np.arange(len(item_mean.index)))
plt.ylabel('MEAN price')
plt.plot(item_mean.values)
```

```python
item_bar = fileData.groupby('Outlet_Size').Item_Outlet_Sales.mean()
item_bar.sort_values(inplace= True)
x = item_bar.index
y = item_bar.values
plt.xlabel('Outlet Size')
plt.ylabel('Sales')
plt.title('MEAN sales for each outlet size')
# plt.xticks(labels=x,ticks=np.arange(len(x)))
# plt.plot(x,y)
plt.bar(x,y,color=['yellow','red','blue'])
```

```python
plt.title('Distribution of Item price')
plt.hist(fileData['Item_MRP'], bins=20, color='lightblue')
```

```python
plt.title('Boxplots: Distribution of Sales')
red_diamond = dict(markerfacecolor='r', marker='D')
plt.boxplot(fileData['Item_Outlet_Sales'], flierprops=red_diamond)
```
